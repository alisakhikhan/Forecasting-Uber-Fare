# Refining the Locations to USA (NY) Only!

As the map below shows (sourced from Tableau), the longitudes and latitudes are very different. We have some locations in the ocean (probably mistakes), and some in Europe, Columbia, and NY/NJ area. This raises concerns because the different areas have different demands and pricing strategies. For example, the culture in the US is very different from Spain, where meal times are much later (and thus, demand and pricing for later Ubers may be higher in Spain, contradicting lower fares at this same time in the USA).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Assume DataFrame is already loaded into variable `df`
df = pd.read_csv('uber_clean.csv')
```

```python
df.head()
```

On Tableau we hovered over the furthest points on the extremities for the USA and then filtered the df to include only these points:


```python
df = df[df['Dropoff Longitude'] <= -69]
```

```python
df = df[df['Dropoff Latitude'] >= -38]
```
