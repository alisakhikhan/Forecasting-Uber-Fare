# Ride Share App Data, Preparation

```python
import pandas as pd
df = pd.read_csv('uber.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0','key'], axis=1, inplace=True)
df.head()
```

```python
df.pickup_datetime=pd.to_datetime(df.pickup_datetime)
```

```python
df['year'] = df.pickup_datetime.dt.year
df['month'] = df.pickup_datetime.dt.month
df['weekday'] = df.pickup_datetime.dt.weekday
df['hour'] = df.pickup_datetime.dt.hour
```

```python
df.head()
```

2015-05-07 was a Thusday (from looking up the date on google) and it is coded to be weekday 3, so we know that:
- Monday = 0
- Tuesday = 1
- Wednesday = 2
- Thursday = 3
- Friday = 4
- Saturday = 5
- Sunday = 6

```python
df['pickup_date'] = df['pickup_datetime'].dt.date
df['pickup_time'] = df['pickup_datetime'].dt.time
df
```

```python
df['am_pm'] = df['pickup_datetime'].dt.hour.apply(lambda x: 'AM' if x < 12 else 'PM')
df
```

```python
df.drop('pickup_datetime', axis=1, inplace=True)
```

```python
new_column_names = {
    'fare_amount': 'Fare Amount ($)',
    'pickup_longitude': 'Pickup Longitude',
    'pickup_latitude': 'Pickup Latitude',
    'dropoff_longitude': 'Dropoff Longitude',
    'dropoff_latitude': 'Dropoff Latitude',
    'passenger_count': 'Number of Passengers',
    'year': 'Year',
    'month': 'Month',
    'weekday': 'Day of Week',
    'hour': 'Hour',
    'pickup_date': 'Pick Up Date',
    'pickup_time': 'Pick Up Time',
    'am_pm': 'AM/PM'

}

# Rename the columns
df.rename(columns=new_column_names, inplace=True)
```

```python
df
```

```python
df.isnull().values.any()
missing_vals = len(df.isnull().sum())
```

```python
percentage_missing = missing_vals/len(df) * 100
percentage_missing
```

```python
df.dropna(inplace=True)
len(df)
```
